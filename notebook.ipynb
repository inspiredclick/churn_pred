```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, roc_auc_score, classification_report, confusion_matrix
from sklearn.utils import shuffle
```

```python
df['Churn'] = np.where(df['EndDate'].isnull(), 0, 1)
df.head()
```

Above we assume if the end date is set then the customer has churned. We store this value in a new 'Churn' column and this will become our target for our ML algorithm. Below we one-hot encode the data for several columns. These are categorical columns and one-hot encoding will help the machine learning libraries form better predictions on our data. 

```python
df = pd.get_dummies(df, columns=['InternetService', 'Type', 'PaymentMethod', 'gender'], drop_first=True)
pd.set_option('display.max_columns', None)  
print(df.head())
```

Next we split our data into a test, train, and validation set. 

```python
target = df['Churn']
features = df.drop(columns=['customerID', 'Churn', 'BeginDate', 'EndDate'], axis=1)

features_train, features_temp, target_train, target_temp = train_test_split(features, target, test_size=0.25, random_state=12345)

features_valid, features_test, target_valid, target_test = train_test_split(features_temp, target_temp, test_size=0.5, random_state=12345)
```

After that we scale our numerical columns. Scaling is important for 'MonthlyCharges' and 'TotalCharges' columns because the large number can alter the performance of several models we will use. By standardizing the range of the columns it provides a clearer signal to learn on.

```python
numeric_cols = ['MonthlyCharges', 'TotalCharges']

scaler = StandardScaler()

features_train = features_train.copy()
features_valid = features_valid.copy()
features_test = features_test.copy()

features_train[numeric_cols] = scaler.fit_transform(features_train[numeric_cols])
features_valid[numeric_cols] = scaler.transform(features_valid[numeric_cols])
features_test[numeric_cols] = scaler.transform(features_test[numeric_cols])

display(features_train.head())
```

## First Model

The first model we will try is the DecisionTreeClassifier model. This model is great for classification and regression. It creates simple rules for the data features. 

```python
best_depth = 0
best_f1 = 0
best_roc_auc = 0
for depth in range(1,10):
    model = DecisionTreeClassifier(random_state=12345, max_depth=depth)
    model.fit(features_train, target_train)
    predicted_valid = model.predict(features_valid)
    probabilities_valid = model.predict_proba(features_test)
    probabilities_one_valid = probabilities_valid[:, 1]

    if f1_score(target_valid, predicted_valid) > best_f1:
        best_f1 = f1_score(target_valid, predicted_valid)
        best_depth = depth
        best_roc_auc = roc_auc_score(target_valid, probabilities_one_valid)

print(f"Best max_depth = {best_depth} f1 score {best_f1} roc_auc {best_roc_auc}")
```

```python
model = DecisionTreeClassifier(random_state=12345, max_depth=6)
model.fit(features_train, target_train)
predicted_valid = model.predict(features_valid)
print(classification_report(target_valid, predicted_valid))
probabilities_valid = model.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]
roc_auc = roc_auc_score(target_valid, probabilities_one_valid)
print(f"ROC AUC: {roc_auc}")
```

```python
model = DecisionTreeClassifier(random_state=12345, max_depth=6, class_weight='balanced')
model.fit(features_train, target_train)
predicted_valid = model.predict(features_valid)
print(classification_report(target_valid, predicted_valid))
probabilities_valid = model.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]
roc_auc = roc_auc_score(target_valid, probabilities_one_valid)
print(f"ROC AUC: {roc_auc}")
```

We have some good performance but we will try tuning the model even more. Below we are checking the balance of classes. As we can see a class imabalance exists. Below we will define functions to correct imbalance.

```python
# CONFUSION MATRIX
result = confusion_matrix(target_valid, predicted_valid)
print(result)
```

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)

    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345
    )

    return features_upsampled, target_upsampled

def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)]
        + [features_ones]
    )
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)]
        + [target_ones]
    )

    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345
    )

    return features_downsampled, target_downsampled
```

```python
model = DecisionTreeClassifier(random_state=12345, max_depth=6, class_weight='balanced')

best_f1 = 0
best_repeat = 0
best_roc_auc = 0
for repeat in range(1,11):
    features_upsampled, target_upsampled = upsample(
        features_train, target_train, repeat
    )
    
    model.fit(features_upsampled, target_upsampled)
    predicted_valid = model.predict(features_valid)

    probabilities_valid = model.predict_proba(features_valid)
    probabilities_one_valid = probabilities_valid[:, 1]

    f1 = f1_score(target_valid, predicted_valid)
    if f1 > best_f1:
        best_f1 = f1
        best_repeat = repeat
        best_roc_auc = roc_auc_score(target_valid, probabilities_one_valid)
print('Repeat:', best_repeat)
print('F1:', best_f1)
print('ROC AUC:', best_roc_auc)
```

```python
best_f1 = 0
best_repeat = 0
best_roc_auc = 0
for repeat in range(1,6):
    features_downsampled, target_downsampled = downsample(
        features_train, target_train, (repeat/10)
    )

    model.fit(features_downsampled, target_downsampled)
    predicted_valid = model.predict(features_valid)

    probabilities_valid = model.predict_proba(features_valid)
    probabilities_one_valid = probabilities_valid[:, 1]

    f1 = f1_score(target_valid, predicted_valid)
    if f1 > best_f1:
        best_f1 = f1
        best_repeat = repeat
        best_roc_auc = roc_auc_score(target_valid, probabilities_one_valid)

print('Fraction:', best_repeat/10)
print('F1:', best_f1)
print('ROC AUC:', best_roc_auc)
```

# Conclusion

In this project I explored several machine learning algorithms to predict the churn of Interconnects customers. Predicting churn for Interconnect is important to increase revenue and also have the ability to message customers before they leave the service. This also improves their retention rate. 

The first step in this process is to load and clean up the data. First the csv data is loaded and joined on the customerId column. Next we clean up the data such as the date columns. These columns were in an object format and converting them to datetime format makes this data easier to work with. The category columns (ex 'MultipleLines', 'Partner', 'Dependents', etc) can be converted into true/false columns. At this point a new column can be created called Churn. The value (true/false) is based on whether the end date is null. A null end date indicates the customer has not left Interconnect. This will be the target column. Finally we can get dummies for several category columns with multiple categories. This transforms text based categories into numerical values.

Now the data is prepared, the next phase of the project can begin. First the target column Churn is separated into a target and the rest is used as features. Several columns are also dropped ('customerID', 'Churn', 'BeginDate', 'EndDate') that do not provide us any value. Here the feature and target data is split into test, train, and validation sets. There are also two columns 'MonthlyCharges', 'TotalCharges' that are scaled. Scaling data allows us to flatten out any outliers and noise in our data. 

DecisionTreeClassifier is the first attempt at solving Interconnects churn prediction problem. First step is to find the optimal max depth for the tree. In this case the max depth is 6 and produces an f1 score of 0.79 and a AUC-ROC score of 0.82. This potentially can be improved on by upsampling or downsampling the data. Upsampling or downsampling alters the data by increasing the number of data points to create a stronger signal or removes data points to create less noise. Using the confusion matrix function there is an imbalance in classes and changing the sampling rate can improve the performance. For the DecisionTreeClassifier upsampling/downsampling only creates a small increase in performance. 

The second model is LogisticRegression. Here the same steps are performed. First the base line is established and produces an f1 score of 0.74 and AUC-ROC 0.54. Upsampling and downsampling does not improve the model and the base line results can be improved upon.

The third algorithm tried is the RandomForestClassifier algorithm. RandomForestClassifier has more hyperparameters to tune but follows a similar pattern to the previous algorithms. The hyperparameters that are being optimized are the n_estimators and max_depth. The f1 score is 0.57 and the AUC-ROC score 0.84. If the classes are balanced using the class_weight hyperparameter then the scores improve slightly with an f1 score of 0.65 roc_auc 0.84. Next the hyperparameters min_samples_split and min_samples_leaf can be tuned. Similar to the other algorithms, changing the sampling rate may improve the performance. Combining the hyperparameter tuning and downsampling produces good results. The end result is a f1 score of 0.77 and a AUC-ROC score of 0.85 which is the highest score. 
This project was challenging in that there are many ways to solve this problem and only by experimenting will the solution be discovered. The end result is that a solid solution has been discovered that Interconnect can use to identify what users are about to leave the service.
